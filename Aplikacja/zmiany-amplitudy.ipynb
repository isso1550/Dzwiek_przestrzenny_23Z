```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write
from IPython.display import Audio
from numpy.fft import fft, ifft
import wave
from scipy import signal
import glob
```

```python
def verifySignal(filename):    
    obj = wave.open(filename, "rb")
    #print(obj.getparams())
    nchannels = obj.getnchannels()
    framerate = obj.getframerate()
    frames = obj.readframes(-1)
    fr = np.frombuffer(frames, dtype=np.int16)
    left = fr[::2]
    right = fr[1::2]
    aoa = int(filename.split("e")[-1].split("a")[0])
    lamp = max(left)
    ramp = max(right)
    return aoa, lamp, ramp, lamp/(lamp+ramp), ramp/(lamp+ramp)
```

```python
verifySignal("./diffuse/elev0/H0e000a.wav")
```

```python
for filename in glob.glob("./diffuse/elev0/*"):
    print(verifySignal(filename))
```
