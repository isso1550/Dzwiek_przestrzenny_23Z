```python
import numpy  as np
import matplotlib.pyplot as plt

outfile = "../Dokumenty/Materialy/Wykresy/LPAmplitudy.png"

angles = np.arange(0, np.pi+0.01, np.pi/12)

l = [1-angle/np.pi for angle in angles]
r = [angle/np.pi for angle in angles]
fig = plt.figure()
plt.plot(np.rad2deg(angles), l, color='green', label = "Lewy kanal")
plt.plot(np.rad2deg(angles), r, color='red', label = "Prawy kanal")
plt.plot(np.rad2deg(angles), np.add(l,r), color='orange', label = "Suma kanalw")
ax = plt.gca()
ax.xaxis.label.set_text("Kt midzy rdlem, a odbiornikiem")
ax.yaxis.label.set_text("Mnonik amplitudy")
plt.title("Linear Panning")
plt.legend()
plt.show()
fig.savefig(outfile)
```

```python
import numpy  as np
import matplotlib.pyplot as plt

outfile = "../Dokumenty/Materialy/Wykresy/CPPAmplitudy.png"

angles = np.arange(0, np.pi+0.01, np.pi/12)
l = [np.cos(angle/2) for angle in angles]
r = [np.sin(angle/2) for angle in angles]
fig = plt.figure()
plt.plot(np.rad2deg(angles), l, color='green', label = "Lewy kanal")
plt.plot(np.rad2deg(angles), r, color='red', label = "Prawy kanal")
plt.plot(np.rad2deg(angles), np.add(l,r), color='orange', label = "Suma kanalw")
ax = plt.gca()
ax.xaxis.label.set_text("Kt midzy rdlem, a odbiornikiem")
ax.yaxis.label.set_text("Mnonik amplitudy")
plt.title("Constant Power Panning")
plt.legend()
plt.show()
fig.savefig(outfile)
```
